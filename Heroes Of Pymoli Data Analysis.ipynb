```python
# delete duplicates/group by gender
Gengroup = purchase_data.groupby(["Gender"])


totlpurchases = Gengroup["Purchase ID"].count()
purchPrice = Gengroup["Price"].mean()
purchValue = Gengroup["Price"].sum()


# Find normalized data
Avg = (Gengroup["Price"].sum() / Gengroup["Purchase ID"].count())

# Create new DataFrame
Purchase_df= pd.DataFrame({"Purchase Count": totalpurchases,
                              "Average Purchase Price": purchPrice,
                              "Total Purchase Value": purchValue,
                              "Avg Total Purchase per Person": Avg})

Purchase_df
```
